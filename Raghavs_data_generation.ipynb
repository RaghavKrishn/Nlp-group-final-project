```python
import sympy as sm
import random as rand
import json

def random_generator():
    return rand.uniform(0.1, 2.0)

def heat_data(iter):
    list_data = []
    x, t = sm.symbols("x t")
    u = sm.Function("u")
    u_t = sm.diff(u(x, t), t)
    u_xx = sm.diff(u(x, t), x, 2)

    for i in range(iter):
        a = round(random_generator(), 2)
        reasoning = (
            f"The equation contains a 1st-order time derivative (u_t) on the left hand side "
            f"and a 2nd-order spatial derivative (u_xx) on the right hand side. "
            f"The coefficient alpha = {a} scales the spatial diffusion term. "
            f"This structure is characteristic of diffusive transport — spatial curvature "
            f"drives temporal change. The equation belongs to the Heat/Diffusion family."
        )
        equation = sm.Eq(u_t, a * u_xx)
        instance = {
            "family": "Heat",
            "coefficients": {"alpha": a},
            "dialects": {
                "latex": sm.latex(equation),
                "prefix": f"= d(u,t) *({a} d(d(u,x),x))",
                "postfix": f"u t d {a} u x x d d * =",
                "natural": f"The time derivative of u equals {a} times the second spatial derivative of u."
            },
            "labels": {
                "behavioral": "Heat",
                "operators": ["exp", "polynomial"],
                "reasoning": reasoning
            }
        }
        list_data.append(instance)
    return list_data

def wave_data(iter):
    list_data = []
    x, t = sm.symbols("x t")
    u = sm.Function("u")
    u_tt = sm.diff(u(x, t), t, 2)
    u_xx = sm.diff(u(x, t), x, 2)

    for i in range(iter):
        c = round(random_generator(), 2)
        c2 = round(c**2, 2)
        reasoning = (
            f"The equation contains a 2nd-order time derivative (u_tt) on the left hand side "
            f"and a 2nd-order spatial derivative (u_xx) on the right hand side. "
            f"The coefficient c^2 = {c2} scales the spatial wave term. "
            f"This structure is characteristic of wave propagation. "
            f"The equation belongs to the Wave family."
        )
        equation = sm.Eq(u_tt, c**2 * u_xx)
        instance = {
            "family": "Wave",
            "coefficients": {"c": c},
            "dialects": {
                "latex": sm.latex(equation),
                "prefix": f"= d(d(u,t),t) *({c2} d(d(u,x),x))",
                "postfix": f"u t t d d {c2} u x x d d * =",
                "natural": f"The second time derivative of u equals {c2} times the second spatial derivative of u."
            },
            "labels": {
                "behavioral": "Wave",
                "operators": ["sin", "cos", "polynomial"],
                "reasoning": reasoning
            }
        }
        list_data.append(instance)
    return list_data

def burgers_data(iter):
    list_data = []
    x, t = sm.symbols("x t")
    u = sm.Function("u")
    u_t = sm.diff(u(x, t), t)
    u_x = sm.diff(u(x, t), x)
    u_xx = sm.diff(u(x, t), x, 2)

    for i in range(iter):
        nu = round(random_generator(), 2)
        reasoning = (
            f"The equation contains a 1st-order time derivative (u_t) and a nonlinear "
            f"convective term (u * u_x) on the left hand side, and a 2nd-order spatial "
            f"derivative (u_xx) scaled by viscosity nu = {nu} on the right hand side. "
            f"The nonlinear convection balanced against diffusion is characteristic of "
            f"the Burgers equation family."
        )
        equation = sm.Eq(u_t + u(x, t) * u_x, nu * u_xx)
        instance = {
            "family": "Burgers",
            "coefficients": {"nu": nu},
            "dialects": {
                "latex": sm.latex(equation),
                "prefix": f"= +(d(u,t) *(u d(u,x))) *({nu} d(d(u,x),x))",
                "postfix": f"u t d u u x d * + {nu} u x x d d * =",
                "natural": f"The time derivative of u plus u times the spatial derivative of u equals {nu} times the second spatial derivative of u."
            },
            "labels": {
                "behavioral": "Burgers",
                "operators": ["tanh", "polynomial"],
                "reasoning": reasoning
            }
        }
        list_data.append(instance)
    return list_data

def laplace_data(iter):
    list_data = []
    x, y = sm.symbols("x y")
    u = sm.Function("u")
    u_xx = sm.diff(u(x, y), x, 2)
    u_yy = sm.diff(u(x, y), y, 2)

    for i in range(iter):
        k = round(random_generator(), 2)
        reasoning = (
            f"The equation contains a scaled sum of a 2nd-order spatial derivative in x (u_xx) "
            f"and a 2nd-order spatial derivative in y (u_yy), both scaled by k = {k}, set equal to zero. "
            f"This structure is characteristic of steady-state diffusion with no time dependence. "
            f"The equation belongs to the Laplace family."
        )
        equation = sm.Eq(k * (u_xx + u_yy), 0)
        instance = {
            "family": "Laplace",
            "coefficients": {"k": k},
            "dialects": {
                "latex": sm.latex(equation),
                "prefix": f"= *({k} +(d(d(u,x),x) d(d(u,y),y))) 0",
                "postfix": f"{k} u x x d d u y y d d + * 0 =",
                "natural": f"{k} times the sum of the second spatial derivative of u in x and the second spatial derivative of u in y equals zero."
            },
            "labels": {
                "behavioral": "Laplace",
                "operators": ["sin", "cos", "exp", "polynomial"],
                "reasoning": reasoning
            }
        }
        list_data.append(instance)
    return list_data

def advection_data(iter):
    list_data = []
    x, t = sm.symbols("x t")
    u = sm.Function("u")
    u_t = sm.diff(u(x, t), t)
    u_x = sm.diff(u(x, t), x)

    for i in range(iter):
        c = round(random_generator(), 2)
        reasoning = (
            f"The equation contains a 1st-order time derivative (u_t) and a 1st-order "
            f"spatial derivative (u_x) scaled by advection speed c = {c}, summed to zero. "
            f"This structure is characteristic of pure transport — the solution propagates "
            f"at constant speed without diffusion. The equation belongs to the Advection family."
        )
        equation = sm.Eq(u_t + c * u_x, 0)
        instance = {
            "family": "Advection",
            "coefficients": {"c": c},
            "dialects": {
                "latex": sm.latex(equation),
                "prefix": f"= +(d(u,t) *({c} d(u,x))) 0",
                "postfix": f"u t d {c} u x d * + 0 =",
                "natural": f"The time derivative of u plus {c} times the spatial derivative of u equals zero."
            },
            "labels": {
                "behavioral": "Advection",
                "operators": ["exp", "polynomial"],
                "reasoning": reasoning
            }
        }
        list_data.append(instance)
    return list_data

def main():
    dataset = []
    dataset.extend(heat_data(2000))
    dataset.extend(wave_data(2000))
    dataset.extend(burgers_data(2000))
    dataset.extend(laplace_data(2000))
    dataset.extend(advection_data(2000))

    rand.shuffle(dataset)

    with open("dataset.jsonl", "w") as f:
        for instance in dataset:
            f.write(json.dumps(instance) + "\n")

    print(f"Written {len(dataset)} instances to dataset.jsonl")

main()
```
